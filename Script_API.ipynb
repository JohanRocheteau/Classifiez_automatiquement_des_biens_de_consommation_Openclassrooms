```python
df = pd.DataFrame()

for i in (20,40,60,80,100,120,140,160,180,200) : 
    
    # Liens : 
    link = 'https://api.edamam.com/api/food-database/v2/parser?session={}&app_id=85b409cf&app_key=69d004c3e3c5b967261e625baea627fc'.format(i)

    headers = {'X-RapidAPI-Key': 'db90e9ef61mshe34c06a99cbdf81p1ba79bjsn3deaf2e05890',
        'X-RapidAPI-Host': 'edamam-food-and-grocery-database.p.rapidapi.com'}

    params = {'ingr': 'champagne'}
    
    # Récupération des jsons :
    r = requests.get(link.format(i), headers = headers, params = params)
    
    # Transformation des json en données :
    data = r.json()
    data = data['hints']
    Data = pd.DataFrame(data)
    
    # Récupération des données :
    for i in range(len(Data)):
        donnee = pd.DataFrame(index = [0])
        donnee['foodId'] = Data['food'][i]['foodId']
        donnee['label'] = Data['food'][i]['label']
        donnee['category'] = Data['food'][i]['category']
        try:
            donnee['foodContentsLabel'] = Data['food'][i]['foodContentsLabel']
        except:
            donnee['foodContentsLabel'] = np.nan
        try :
            donnee['image'] = Data['food'][i]['image']
        except :
            donnee['image'] = np.nan
        df = pd.concat([df,donnee])
```
